```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('hwc.csv')
df_habitable = df[df['P_HABITABLE'] == 2]
df_casihabitable = df[df['P_HABITABLE'] == 1]
df_nohabitable = df[df['P_HABITABLE'] == 0]
```

# Pregunta Guía

¿Qué combinación de masa, radio y densidad hace más probable que un planeta retenga atmósfera y
agua líquida? El foco está en las propiedades físicas que determinan si un planeta puede mantener
condiciones superficiales estables

```python
df_habitable
```

## ¿Cómo se distribuyen los tipos planetarios en el espacio masa–radio?

```python
plt.scatter(df_habitable['P_MASS'], df_habitable['P_RADIUS'], s= 6, color='green', label="Habitabilidad=2")
plt.scatter(df_casihabitable['P_MASS'], df_casihabitable['P_RADIUS'], s= 6, color='red', label="Habitabilidad=1")
plt.scatter(df_nohabitable['P_MASS'], df_nohabitable['P_RADIUS'], s= 6, alpha=0.01, color='grey',label="Habitabilidad=0")

plt.xlabel('Masa')
plt.ylabel('Radio')
plt.xlim(0,12)
plt.ylim(0,3.5)
plt.title('Masa - Radio')
plt.legend()
plt.grid()
plt.show()
```

## ¿Existe relación entre densidad media y velocidad de escape?

```python
plt.scatter(df_habitable['P_DENSITY'], df_habitable['P_ESCAPE'], s=4, alpha=1, color="green")
plt.scatter(df["P_DENSITY"], df["P_ESCAPE"], s=4, alpha=0.05, color="grey")
plt.xlim(-1,2)
plt.ylim(-1,30)
plt.xlabel('Densidad')
plt.ylabel('Velocidad de Escape')
```

## ¿Qué combinación de masa y temperatura parece más favorable para retener atmósfera?

```python
plt.scatter(df_habitable['P_MASS'], df_habitable['P_TEMP_EQUIL'], s=4, alpha=1, color="green")
plt.scatter(df["P_MASS"], df["P_TEMP_EQUIL"], s=4, alpha=0.05, color="grey")
plt.xlim(0,20)
plt.ylim(0,2000)
plt.xlabel('Densidad')
plt.ylabel('Velocidad de Escape')
```

## Condición de habitabilidad de base a la gravedad, cercana a la tierra.

```python
df_ideales = df_habitable[df['P_GRAVITY'] <= 1.06]
df_ideales
```

```python
plt.scatter(df['P_MASS'], df['P_RADIUS'],s= 8, alpha= 0.1,color='grey')
plt.scatter(df_ideales['P_MASS'], df_ideales['P_RADIUS'],s= 8, alpha= 1,color='red', label="Candidatos")
plt.xlim(0,50)
plt.ylim(0,10)
plt.xlabel('Masa')
plt.ylabel('Radio')
plt.title('Masa - Radio')
plt.grid()
plt.legend()
plt.show()

```
